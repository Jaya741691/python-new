```python
from functools import reduce
l=["hello","hii","how","are","you?"]
def fun2(x):
    s=""
    for i in x:
        if i not in "aeiouAEIOU":
            s+=i
    return s
k=list(map(fun2,l))
print(k)
# j=list(map(lambda x: reduce(lambda y,z: y + ord(z),x,0),k))
def fun3(y):
    s=0
    for i in y:
        s+=ord(i)
    return s
j=list(map(fun3,k))

print(j)
```

```python
#  PARAMETERS + LAMBDA: Write a function apply_operation(a, b, op) where op is a
# lambda. Call it with operations for add, subtract, and multiply.
a=20
b=30
o=input()
def oper(a,b,op):
    print(op(a,b))
if o=="+":
    oper(a,b,lambda x,y:x+y)
elif o=="-":
    oper(a,b,lambda x,y:x-y)
elif o=="*":
    oper(a,b,lambda x,y:x*y)
```

```python
#  *args + RECURSION: Write a recursive function that takes *args of numbers and
# returns their sum WITHOUT using the built-in sum().

def sum1(*a):
    s=0
    for i in a:
        s+=i
    return s
print(sum1(1,2,3,4,5,6))
```

```python
#  DEFAULT + KEYWORD + LAMBDA: Write a function make_greeting(name,
# prefix='Hello', formatter=lambda x: x) that applies formatter to the final greeting string. Test
# with str.upper as the formatter.

def form(name,format,prefix="Hello"):
    k=prefix+ " " + name
    return format(k)
print(form("jaya",lambda x:x.upper()))
```
