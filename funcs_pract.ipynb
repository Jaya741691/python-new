```python
def mul_all(*args):
    res=1
    for i in args:
        res*=i
    return res
print(mul_all(10,20,3,4,3))
```

```python
def display_tags(**kwargs):
    for i,j in kwargs.items():
        print(i,":",j)
display_tags(name='jaya',age=20,gender='male')
```

```python
def fun(func,var):
    return func(func(func(var)))
def add(x):
    return x+5
print(fun(add,10))
```

```python
def flatten(l,dpt):
    if dpt==0:
        return l
    r=[]
    for i in l:
        if isinstance(i,list):
            r.extend(flatten(i,dpt-1))

        else:
            r.append(i)
    return r
l=[1,[2,3,[4,[5,[6]]],5],6,[6]]
print(flatten(l,1))
print(flatten(l,2))
print(flatten(l,3))
print(flatten(l,4))
```

```python
def sumc(*args):
    if not args:
        return 0
    return args[0]+ sumc(*args[1:])
print(sumc(1,2,3))
```
