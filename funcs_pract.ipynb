```python
def mul_all(*args):
    res=1
    for i in args:
        res*=i
    return res
print(mul_all(10,20,3,4,3))
```

```python
def display_tags(**kwargs):
    for i,j in kwargs.items():
        print(i,":",j)
display_tags(name='jaya',age=20,gender='male')
```

```python
def fun(func,var):
    return func(func(func(var)))
def add(x):
    return x+5
print(fun(add,10))
```
