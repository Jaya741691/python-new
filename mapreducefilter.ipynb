```python
l = [[1, 2], [3, 4], [5, 6]]
k=list(map(lambda x:list(map(lambda y:y+5,x)),l))
print(k)
```

```python
d = {"apple": 100, "banana": 40, "cherry": 150}
k=dict(filter(lambda x: x[1] > 50,d.items()))
print(k)
```

```python
from functools import reduce
l=[1,2,3,4,5,5,3,5,3,34,3]
k=reduce(lambda x,y:x if x>y else y,l)
print(k)
```

```python
s="jayasimha"
k=list(map(lambda x:ord(x),s))
print(k)
```

```python
from functools import reduce
l=[1,5,3,6,20,13]
c=reduce(lambda x,y:x+y,list(map(lambda x:x,filter(lambda x:x%5==0,list(map(lambda x:x**2,l))))))
print(c)
```
